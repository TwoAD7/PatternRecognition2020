```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
p_bill = 45
p_tip = 1
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.plot(p_bill,p_tip,'gs')
```

```python
tips['dist2'] = (tips.total_bill - p_bill)**2 + (tips.tip - p_tip)**2
```

```python
tips.sort_values(by=['dist2'],inplace=True)
```

```python
tips.iloc[:5]
```

These are the closest 5 neighbors!

Since 4 of 5 of these are smokers, then our point is 80% likely to be a smoker too

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.plot(p_bill,p_tip,'gs')
sns.scatterplot(x="total_bill", y="tip",data=tips.iloc[:5], marker='o', s=200,facecolors="none", edgecolor="black")
```

```python
tips['smoker'].unique()
```

```python
tips['smoker_class'] = pd.to_numeric(tips['smoker'].map({"Yes":1, "No":0}))
```
